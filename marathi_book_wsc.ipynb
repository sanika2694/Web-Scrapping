```python
import requests
from bs4 import BeautifulSoup

# URL of the Cricbuzz page you want to scrape
url = "https://www.cricbuzz.com/cricket-match/live-scores"

# Send a request to the website
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.text, "html.parser")

# Extract data (modify selectors based on the page's structure)
matches = soup.find_all("div", class_="cb-mtch-lst")
for match in matches:
    title = match.find("a").text
    # status = match.find("div", class_="cb-text-live").text
    print(f"Match: {title}")
    # print(f"Status: {status}")
```
